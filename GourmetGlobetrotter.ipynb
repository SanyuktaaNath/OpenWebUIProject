Salt and pepper, to taste
For the tzatziki sauce:

1 cup Greek yogurt
1/2 cup diced cucumber
1/4 cup chopped fresh dill
1 tablespoon lemon juice
Salt and pepper, to taste
For the assembly:

4-6 pita breads
Sliced tomatoes, onions, and cucumbers (for garnish)
Crispy pickled turnips (see below for recipe)
To make the crispy pickled turnips:

1 large turnip, peeled and thinly sliced
1/2 cup white vinegar
1/4 cup sugar
1 tablespoon salt
1/4 teaspoon black pepper
Instructions:

Add the chicken to the marinade and mix well. Cover the bowl with plastic wrap and refrigerate for at least 30 minutes or up to 2 hours.
Preheat a grill or grill pan to medium-high heat. Remove the chicken from the marinade and cook for 5-7 minutes per side, or until cooked through.

Warm the pita breads by wrapping them in foil and heating them in the oven for a few minutes.
